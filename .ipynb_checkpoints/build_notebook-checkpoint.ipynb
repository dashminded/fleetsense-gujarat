```
fig = px.bar(rev, x='month', y='total_revenue_inr', color=rev['year'].astype(str),
             barmode='group', title='💰 Monthly Revenue: 2023 vs 2024',
             labels={'total_revenue_inr':'Revenue (₹)','month':'Month','color':'Year'},
             template='plotly_dark')
fig.update_layout(xaxis=dict(tickmode='linear'))
fig.show()
print("💡 Insight: Revenue grew YoY, with October–November showing the largest jump due to Navratri-Diwali clustering.")
""")

code("""# Chart 5 🚌 — Avg Profit & Occupancy by Bus Type
bus_stats = (df.groupby('bus_type')
               .agg(avg_profit=('profit_loss_inr','mean'),
                    avg_occ=('occupancy_rate','mean'))
               .reset_index())
```

```
plt.suptitle('🚌 Bus Type Performance', fontsize=14); plt.tight_layout(); plt.show()
print("💡 Insight: AC Sleeper buses yield the highest per-trip profit despite lower seat count, due to premium pricing.")
""")

code("""# Chart 6 🥧 — Demand Label Distribution
demand_counts = df['demand_label'].value_counts().reset_index()
demand_counts.columns = ['demand_label','count']
fig = px.pie(demand_counts, names='demand_label', values='count',
             title='🥧 Demand Label Distribution',
             color_discrete_sequence=px.colors.sequential.RdYlGn,
             template='plotly_dark')
fig.show()
print("💡 Insight: ~62% of trips fall in Normal demand — only ~8% reach Overcrowded, signalling untapped revenue during peak days.")
""")

code("""# Chart 7 📦 — Profitable vs Loss-Making Trips per Route (Stacked Bar)
df['profitable'] = df['profit_loss_inr'] >= 0
stack = (df.groupby(['route_name','profitable'])
           .size().unstack(fill_value=0)
           .rename(columns={True:'Profitable',False:'Loss-Making'}))
```

```
stack.plot(kind='bar', stacked=True, figsize=(13,5),
           color=['#06d6a0','#ef233c'], edgecolor='none')
plt.title('📦 Profitable vs Loss-Making Trips per Route', fontsize=13)
plt.xlabel('Route'); plt.ylabel('Trip Count')
plt.xticks(rotation=30, ha='right'); plt.legend(); plt.tight_layout(); plt.show()
print("💡 Insight: Tier-3 routes like AMD→Dwarka have higher loss-making trip ratios — consider dynamic pricing or reduced frequency.")
""")

# ── Section 3 ───────────────────────────────────────────────────────────────
md("---\n## ⚙️ Section 3 · Feature Engineering")

code("""# 1. is_summer_vacation — school summer break drives leisure travel
df['is_summer_vacation'] = (
    ((df['date'].dt.month == 5) & (df['date'].dt.day >= 15)) |
    ((df['date'].dt.month == 6) & (df['date'].dt.day <= 15))
).astype(int)
```

```
# 2. is_long_weekend — Fri or Mon adjacent to weekend increases travel intent
df['is_long_weekend'] = (
    (df['day_of_week'].isin(['Friday','Monday'])) &
    (df['is_weekend'] == 1)
).astype(int)
```

```
# 3. rolling_7d_occ — smoothed demand signal per route+bus combo
df = df.sort_values(['route_id','bus_id','date'])
df['rolling_7d_occ'] = (
    df.groupby(['route_id','bus_id'])['occupancy_rate']
      .transform(lambda x: x.rolling(7, min_periods=1).mean())
)
```

```
# 4. revenue_per_seat — revenue efficiency per available seat
df['revenue_per_seat'] = df['total_revenue_inr'] / df['total_seats']
```

```
# 5. cost_efficiency_ratio — cost as fraction of revenue (<1 = profitable)
df['cost_efficiency_ratio'] = df['total_operational_cost_inr'] / df['total_revenue_inr']
```

```
# 6. quarter — seasonal grouping
df['quarter'] = pd.cut(df['date'].dt.month, bins=[0,3,6,9,12],
                       labels=['Q1','Q2','Q3','Q4'])
```

```
print("✅ Feature engineering complete. New columns:")
print(df[['is_summer_vacation','is_long_weekend','rolling_7d_occ',
          'revenue_per_seat','cost_efficiency_ratio','quarter']].head(3))
""")

code("""# Correlation heatmap of engineered features vs occupancy_rate
feat_cols = ['occupancy_rate','is_summer_vacation','is_long_weekend',
             'rolling_7d_occ','revenue_per_seat','cost_efficiency_ratio',
             'is_festival','is_weekend']
corr = df[feat_cols].corr()
```

```
plt.figure(figsize=(9,7))
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', center=0, linewidths=.5)
plt.title('🔗 Feature Correlation Matrix vs Occupancy Rate', fontsize=13)
plt.tight_layout(); plt.show()
print("💡 Insight: rolling_7d_occ and is_festival are the strongest predictors of occupancy_rate.")
""")

# ── Section 4 ───────────────────────────────────────────────────────────────
md("---\n## 🔮 Section 4 · Prophet Demand Forecasting")

code("""# ── Gujarat-specific holiday calendar ────────────────────────────────────
def make_holidays():
    rows = []
    for yr in [2023, 2024]:
        rows += [
            {'holiday':'Uttarayan',       'ds':f'{yr}-01-14', 'lower_window':0, 'upper_window':1},
            {'holiday':'Holi',            'ds':f'{yr}-03-08', 'lower_window':0, 'upper_window':2},
            {'holiday':'Eid',             'ds':f'{yr}-04-21', 'lower_window':0, 'upper_window':2},
            {'holiday':'SummerVacation',  'ds':f'{yr}-05-15', 'lower_window':0, 'upper_window':30},
            {'holiday':'RathYatra',       'ds':f'{yr}-06-20', 'lower_window':0, 'upper_window':1},
            {'holiday':'Janmashtami',     'ds':f'{yr}-08-26', 'lower_window':0, 'upper_window':1},
            {'holiday':'Navratri',        'ds':f'{yr}-10-15', 'lower_window':0, 'upper_window':9},
            {'holiday':'Diwali',          'ds':f'{yr}-11-12', 'lower_window':-1,'upper_window':4},
            {'holiday':'RannUtsav',       'ds':f'{yr}-12-01', 'lower_window':0, 'upper_window':90},
            {'holiday':'NewYear',         'ds':f'{yr}-12-31', 'lower_window':0, 'upper_window':1},
        ]
    return pd.DataFrame(rows).assign(ds=lambda d: pd.to_datetime(d['ds']))
```

```
gujarat_holidays = make_holidays()
print(f"✅ Holiday calendar built: {len(gujarat_holidays)} entries")
gujarat_holidays.head(6)
""")

code("""# ── Forecast loop — all 12 routes ────────────────────────────────────────
FORECAST_DAYS = 90
all_forecasts  = []
```

```
for route in df['route_name'].unique():
    # aggregate to daily avg occupancy for this route
    agg = (df[df['route_name'] == route]
             .groupby('date')['occupancy_rate']
             .mean()
             .reset_index()
             .rename(columns={'date':'ds','occupancy_rate':'y'}))

    m = Prophet(
        yearly_seasonality=True,
        weekly_seasonality=True,
        seasonality_mode='multiplicative',
        changepoint_prior_scale=0.05,
        holidays=gujarat_holidays
    )
    m.fit(agg)

    future   = m.make_future_dataframe(periods=FORECAST_DAYS)
    forecast = m.predict(future)

    # clip occupancy to realistic bounds
    forecast['yhat']       = forecast['yhat'].clip(0.10, 1.0)
    forecast['yhat_lower'] = forecast['yhat_lower'].clip(0.10, 1.0)
    forecast['yhat_upper'] = forecast['yhat_upper'].clip(0.10, 1.0)

    forecast['route_name'] = route
    all_forecasts.append(forecast[['ds','yhat','yhat_lower','yhat_upper','route_name']])
    print(f"  ✅ {route} — forecast done")
```

```
forecast_df = pd.concat(all_forecasts, ignore_index=True)
future_only = forecast_df[forecast_df['ds'] > '2024-12-31'].copy()
print(f"\\n🔮 Total forecast rows (future only): {len(future_only)}")
""")

code("""# ── Visualise AMD→Surat forecast with thresholds ─────────────────────────
surat = forecast_df[forecast_df['route_name'] == 'AMD→Surat'].copy()
```

```
fig = go.Figure([
    go.Scatter(x=surat['ds'], y=surat['yhat_upper'], fill=None,
               mode='lines', line_color='rgba(0,180,220,0.2)', name='Upper CI'),
    go.Scatter(x=surat['ds'], y=surat['yhat_lower'], fill='tonexty',
               mode='lines', line_color='rgba(0,180,220,0.2)', name='Lower CI'),
    go.Scatter(x=surat['ds'], y=surat['yhat'], mode='lines',
               line=dict(color='#00b4dc', width=2), name='Forecast'),
])
fig.add_hline(y=0.80, line_dash='dash', line_color='green',
              annotation_text='Add Bus Threshold (0.80)')
fig.add_hline(y=0.55, line_dash='dash', line_color='red',
              annotation_text='Remove Bus Threshold (0.55)')
fig.update_layout(title='🔮 AMD→Surat — 90-Day Occupancy Forecast',
                  xaxis_title='Date', yaxis_title='Forecasted Occupancy',
                  template='plotly_dark', yaxis_tickformat='.0%')
fig.show()
print("💡 Insight: AMD→Surat is expected to breach the 'add bus' threshold repeatedly in Jan–Mar 2025 (Rann Utsav overlap).")
""")

# ── Section 5 ───────────────────────────────────────────────────────────────
md("---\n## 🚍 Section 5 · Fleet Recommendation Engine")

code("""# ── Historical avg revenue & cost per trip per route ─────────────────────
route_economics = (df.groupby('route_name')
                     .agg(avg_rev_per_bus =('total_revenue_inr','mean'),
                          avg_cost_per_bus=('total_operational_cost_inr','mean'))
                     .reset_index())
```

```
# ── Base buses per route from routes.csv ─────────────────────────────────
base_map = routes.set_index('route_name')['base_buses_per_day'].to_dict()
```

```
def recommend_fleet(occ_rate, base_buses):
    \"\"\"Return (recommended_buses, recommendation_label, extra_buses).\"\"\"
    if occ_rate >= 0.92:
        extra = 2
    elif occ_rate >= 0.80:
        extra = 1
    elif occ_rate < 0.45:
        extra = -1
    else:
        extra = 0
    recommended = max(1, base_buses + extra)
    label = {2:'Add 2 Buses', 1:'Add 1 Bus', 0:'Keep As-Is', -1:'Remove 1 Bus'}[extra]
    return recommended, label, extra
```

```
# ── Apply to forecast rows ────────────────────────────────────────────────
future_only = future_only.copy()
future_only['base_buses'] = future_only['route_name'].map(base_map).fillna(2).astype(int)
```

```
results = future_only.apply(
    lambda r: recommend_fleet(r['yhat'], r['base_buses']), axis=1,
    result_type='expand'
)
results.columns = ['recommended_buses','recommendation','extra_buses']
future_only = pd.concat([future_only, results], axis=1)
```

```
# ── Merge economics ───────────────────────────────────────────────────────
future_only = future_only.merge(route_economics, on='route_name', how='left')
future_only['est_extra_revenue_inr'] = future_only['extra_buses'] * future_only['avg_rev_per_bus']
future_only['est_extra_cost_inr']    = future_only['extra_buses'].abs() * future_only['avg_cost_per_bus']
future_only['est_net_impact_inr']    = future_only['est_extra_revenue_inr'] - future_only['est_extra_cost_inr']
```

```
print("✅ Fleet recommendations applied.")
print(future_only['recommendation'].value_counts())
""")

code("""# ── Top 20 high-demand days (Add Bus days) ────────────────────────────────
add_days = (future_only[future_only['extra_buses'] > 0]
              .sort_values('yhat', ascending=False)
              .head(20)[['ds','route_name','yhat','recommendation','est_net_impact_inr']])
print("🔴 Top 20 High-Demand Days (Add Bus):")
print(add_days.to_string(index=False))
""")

code("""# ── Top 15 cost-saving days (Remove Bus days) ────────────────────────────
remove_days = (future_only[future_only['extra_buses'] < 0]
                 .sort_values('yhat')
                 .head(15)[['ds','route_name','yhat','recommendation','est_extra_cost_inr']])
print("🟢 Top 15 Cost-Saving Days (Remove Bus):")
print(remove_days.to_string(index=False))
""")

code("""# ── 90-Day Heatmap Calendar — All Routes × Forecast Dates ───────────────
pivot = future_only.pivot_table(
    index='route_name', columns=future_only['ds'].dt.strftime('%b %d'),
    values='yhat', aggfunc='mean'
)
```

```
plt.figure(figsize=(28, 7))
sns.heatmap(pivot, cmap='RdYlGn', vmin=0.2, vmax=1.0,
            linewidths=0.1, linecolor='gray',
            cbar_kws={'label':'Forecasted Occupancy'})
plt.title('📅 90-Day Forecasted Occupancy Heatmap — All Routes', fontsize=14)
plt.xlabel('Forecast Date'); plt.ylabel('Route')
plt.xticks(rotation=90, fontsize=7); plt.tight_layout(); plt.show()
print("💡 Insight: Early Jan 2025 shows cluster-wide high demand — ideal for fleet-wide surge deployment.")
""")

# ── Section 6 ───────────────────────────────────────────────────────────────
md("---\n## 💾 Section 6 · Export & Summary")

code("""# ── Build clean export DataFrame ─────────────────────────────────────────
export_cols = ['ds','route_name','yhat','yhat_lower','yhat_upper',
               'base_buses','recommended_buses','recommendation','extra_buses',
               'est_extra_revenue_inr','est_extra_cost_inr','est_net_impact_inr']
export_df = future_only[export_cols].rename(columns={
    'ds'   : 'forecast_date',
    'yhat' : 'forecast_occ_pct',
    'yhat_lower':'lower_bound',
    'yhat_upper':'upper_bound'
})
```

```
out_path = os.path.join(DATA_PATH, 'fleet_forecast_90days.csv')
export_df.to_csv(out_path, index=False)
print(f"✅ Exported → {out_path}  ({len(export_df):,} rows)")
export_df.head(5)
""")

code("""# ── Final Summary Box ─────────────────────────────────────────────────────
add_count    = (future_only['extra_buses'] > 0).sum()
remove_count = (future_only['extra_buses'] < 0).sum()
net_revenue  = future_only['est_net_impact_inr'].sum()
```
